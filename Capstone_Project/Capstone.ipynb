## Replacing the missing data

Here we can just work on replacing the NaN data with mean and later while binning we can work with categorical data.

```python
#casualty_class = Missing data 99259
#code	label
#1	Driver or rider
#2	Passenger
#3	Pedestrian
avg_casualty_class=df['casualty_class'].astype('float').mean(axis=0)
print("casualty_class:", avg_casualty_class)
df['casualty_class'].replace(np.nan, avg_casualty_class, inplace=True)

#sex_of_casualty = Missing data 99259
#code	label
#1	Male
#2	Female
avg_sex_of_casualty=df['sex_of_casualty'].astype('float').mean(axis=0)
print("sex_of_casualty:", avg_sex_of_casualty)
df['sex_of_casualty'].replace(np.nan, avg_sex_of_casualty, inplace=True)

#age_of_casualty = Missing data 99259
#code	label
#0-120	0-120
avg_age_of_casualty=df['age_of_casualty'].astype('float').mean(axis=0)
print("age_of_casualty:", avg_age_of_casualty)
df['age_of_casualty'].replace(np.nan, avg_age_of_casualty, inplace=True)

#age_band_of_casualty = Missing data 99259
#1	0 - 5
#2	6 - 10
#3	11 - 15
#4	16 - 20
#5	21 - 25
#6	26 - 35
#7	36 - 45
#8	46 - 55
#9	56 - 65
#10	66 - 75
#11	Over 75
avg_age_band_of_casualty=df['age_band_of_casualty'].astype('float').mean(axis=0)
print("age_band_of_casualty:", avg_age_band_of_casualty)
df['age_band_of_casualty'].replace(np.nan, avg_age_band_of_casualty, inplace=True)


#casualty_severity = Missing data 99259
#code	label
#1	Fatal
#2	Serious
#3	Slight
avg_casualty_severity=df['casualty_severity'].astype('float').mean(axis=0)
print("casualty_severity:", avg_casualty_severity)
df['casualty_severity'].replace(np.nan, avg_casualty_severity, inplace=True)

#pedestrian_location = Missing data 99259
#0	Not a Pedestrian
#1	Crossing on pedestrian crossing facility
#2	Crossing in zig-zag approach lines
#3	Crossing in zig-zag exit lines
#4	Crossing elsewhere within 50m. of pedestrian crossing
#5	In carriageway, crossing elsewhere
#6	On footway or verge
#7	On refuge, central island or central reservation
#8	In centre of carriageway - not on refuge, island or central reservation
#9	In carriageway, not crossing
#10	Unknown or other
avg_pedestrian_location=df['pedestrian_location'].astype('float').mean(axis=0)
print("pedestrian_location:", avg_pedestrian_location)
df['pedestrian_location'].replace(np.nan, avg_pedestrian_location, inplace=True)


#pedestrian_movement = Missing data 99259
#0	Not a Pedestrian
#1	Crossing from driver's nearside
#2	Crossing from nearside - masked by parked or stationary vehicle
#3	Crossing from driver's offside
#4	Crossing from offside - masked by  parked or stationary vehicle
#5	In carriageway, stationary - not crossing  (standing or playing)
#6	In carriageway, stationary - not crossing  (standing or playing) - masked by parked or stationary vehicle
#7	Walking along in carriageway, facing traffic
#8	Walking along in carriageway, back to traffic
#9	Unknown or other
avg_pedestrian_movement=df['pedestrian_movement'].astype('float').mean(axis=0)
print("pedestrian_movement:", avg_pedestrian_movement)
df['pedestrian_movement'].replace(np.nan, avg_pedestrian_movement, inplace=True)

#car_passenger = Missing data 99259
#code	label
#0	Not car passenger
#1	Front seat passenger
#2	Rear seat passenger
avg_car_passenger=df['casualty_class'].astype('float').mean(axis=0)
print("car_passenger:", avg_car_passenger)
df['car_passenger'].replace(np.nan, avg_car_passenger, inplace=True)

#bus_or_coach_passenger = Missing data 99259
#0	Not a bus or coach passenger
#1	Boarding
#2	Alighting
#3	Standing passenger
#4	Seated passenger
avg_bus_or_coach_passenger=df['bus_or_coach_passenger'].astype('float').mean(axis=0)
print("bus_or_coach_passenger:", avg_bus_or_coach_passenger)
df['bus_or_coach_passenger'].replace(np.nan, avg_bus_or_coach_passenger, inplace=True)

#pedestrian_road_maintenance_worker = Missing data 99259
#0	No / Not applicable
#1	Yes
#2	Not Known
avg_pedestrian_road_maintenance_worker=df['pedestrian_road_maintenance_worker'].astype('float').mean(axis=0)
print("pedestrian_road_maintenance_worker:", avg_pedestrian_road_maintenance_worker)
df['pedestrian_road_maintenance_worker'].replace(np.nan, avg_pedestrian_road_maintenance_worker, inplace=True)

#casualty_type = Missing data 99259
#0	Pedestrian
#1	Cyclist
#2	Motorcycle 50cc and under rider or passenger
#3	Motorcycle 125cc and under rider or passenger
#4	Motorcycle over 125cc and up to 500cc rider or  passenger
#5	Motorcycle over 500cc rider or passenger
#8	Taxi/Private hire car occupant
#9	Car occupant
#10	Minibus (8 - 16 passenger seats) occupant
#11	Bus or coach occupant (17 or more pass seats)
#16	Horse rider
#17	Agricultural vehicle occupant
#18	Tram occupant
#19	Van / Goods vehicle (3.5 tonnes mgw or under) occupant
#20	Goods vehicle (over 3.5t. and under 7.5t.) occupant
#21	Goods vehicle (7.5 tonnes mgw and over) occupant
#22	Mobility scooter rider
#23	Electric motorcycle rider or passenger
#90	Other vehicle occupant
#97	Motorcycle - unknown cc rider or passenger
#98	Goods vehicle (unknown weight) occupant
avg_casualty_type=df['casualty_type'].astype('float').mean(axis=0)
print("casualty_type:", avg_casualty_type)
df['casualty_type'].replace(np.nan, avg_casualty_type, inplace=True)


#casualty_home_area_type = Missing data 99259
#code	label
#1	Urban area
#2	Small town
#3	Rural
avg_casualty_home_area_type=df['casualty_home_area_type'].astype('float').mean(axis=0)
print("casualty_home_area_type:", avg_casualty_home_area_type)
df['casualty_home_area_type'].replace(np.nan, avg_casualty_home_area_type, inplace=True)

#casualty_imd_decile = Missing data 99259
#code	label
#1	Most deprived 10%
#2	More deprived 10-20%
#3	More deprived 20-30%
#4	More deprived 30-40%
#5	More deprived 40-50%
#6	Less deprived 40-50%
#7	Less deprived 30-40%
#8	Less deprived 20-30%
#9	Less deprived 10-20%
#10	Least deprived 10%
avg_casualty_imd_decile=df['casualty_imd_decile'].astype('float').mean(axis=0)
print("casualty_imd_decile:", avg_casualty_imd_decile)
df['casualty_imd_decile'].replace(np.nan, avg_casualty_imd_decile, inplace=True)

```
